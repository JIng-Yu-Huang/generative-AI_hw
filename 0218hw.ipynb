# 畫函數圖形

畫出 $f(x) = \sin(x)$


```python
# prompt: draw a graph y = sin(x)

import numpy as np
import matplotlib.pyplot as plt

# 建立 x 座標的數值範圍
x = np.linspace(0, 10, 400)

# 計算對應的 y 座標值
y = np.sin(x)

# 建立圖形
plt.plot(x, y)

# 建立圖形，並設定顏色為紅色
plt.plot(x, y, color='pink')  # or use 'r' for short

# 設定圖形的標題和軸標籤
plt.title('y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')

# 顯示圖形
plt.show()
```
