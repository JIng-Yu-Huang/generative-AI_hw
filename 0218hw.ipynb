# 畫函數圖形

畫出 $f(x) = \sin(x)$


```python
# prompt: draw a graph y = sin(x)

import numpy as np
import matplotlib.pyplot as plt

# 建立 x 座標的數值範圍
x = np.linspace(0, 10, 400)

# 計算對應的 y 座標值
y = np.sin(x)

# 建立圖形
plt.plot(x, y)

# 建立圖形，並設定顏色為紅色
plt.plot(x, y, color='pink')  # or use 'r' for short

# 設定圖形的標題和軸標籤
plt.title('y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')

# 顯示圖形
plt.show()
```

畫出$f(x) = tan(x)+2$

```python
import numpy as np
import matplotlib.pyplot as plt

# x的數值範圍
x = np.linspace(-5, 5, 400)

# 計算對應的 y 座標值
y = np.tan(x)+2

# 建立圖形
plt.plot(x, y)

# 設定圖形的標題和軸標籤
plt.title('y = tan(x)+2')
plt.xlabel('x')
plt.ylabel('y')

# 顯示圖形
plt.show()
```

畫出$f(x) = 3x^2+7x+9$

```python
import numpy as np
import matplotlib.pyplot as plt

# x的數值範圍
x = np.linspace(-10, 10, 400)

# 計算對應的y值
y = 3 * x**2 + 7 * x + 9

# 建立圖形，顏色設定為綠色
plt.plot(x, y, color='green')

# 圖形的標題、兩軸
plt.title('y = 3x^2 + 7x + 9')
plt.xlabel('x')
plt.ylabel('y')

# 顯示圖形
plt.show()
```
